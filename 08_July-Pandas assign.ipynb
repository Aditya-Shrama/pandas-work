# Q1  1. what was the total revenues generated from binder , pen ,pencil

```python
Rev_binder = df["Revenue"].where(df['Items']=="Binder").sum()
Rev_pen = df["Revenue"].where(df['Items']=="Pen").sum()
Rev_pencil = df["Revenue"].where(df['Items']=="Pencil").sum()
print("Total revenue generated from Binder :",Rev_binder.round(2))
print("Total revenue generated from Pen :",Rev_pen.round(2))
print("Total revenue generated from Pencil :",Rev_pencil.round(2))
```

# 2. what was the generating revenue generate from the central region ,east,west , where the item is pencil.

```python
Rev_Central= df["Revenue"].where((df['Region']=="Central" )& (df['Items']=="Pencil")).sum().round(2)
Rev_East = df["Revenue"].where((df['Region']=="East")& (df['Items']=="Pencil")).sum().round(2)
Rev_West = df["Revenue"].where((df['Region']=="West")& (df['Items']=="Pencil")).sum().round(2)
print("Total revenue generated from Central region by pencil :",Rev_Central)
print("Total revenue generated from Central region by pencil :",Rev_East)
print("Total revenue generated from Central region by pencil :",Rev_West)
```
